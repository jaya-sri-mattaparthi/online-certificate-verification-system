```python
import yaml

yaml_path = "/kaggle/input/datasets/saisameer2457/document/dataset/data.yaml"

with open(yaml_path) as f:
    data = yaml.safe_load(f)

data["train"] = "/kaggle/input/datasets/saisameer2457/document/dataset/train/images"
data["val"]   = "/kaggle/input/datasets/saisameer2457/document/dataset/valid/images"
data["test"] = "/kaggle/input/datasets/saisameer2457/document/dataset/test/images"

with open("data.yaml", "w") as f:
    yaml.dump(data, f)

print(data)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0
)
```

```python
metrics = model.val()

print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```
